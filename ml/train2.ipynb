```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
import torch.nn.functional as F
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
```

```python
class Model(nn.Module):

    def __init__(self, n_observations, n_actions):
        super(Model, self).__init__()
        self.layer1 = nn.Linear(n_observations, 512)
        self.bn1 = nn.BatchNorm1d(512)
        self.layer1d = nn.Dropout(p=0.2)
        self.layer2 = nn.Linear(512, 256)
        self.bn2 = nn.BatchNorm1d(256)
        self.layer2d = nn.Dropout(p=0.2)
        self.layer3 = nn.Linear(256, 128)
        self.bn3 = nn.BatchNorm1d(128)
        self.layer3d = nn.Dropout(p=0.2)
        self.layer4 = nn.Linear(128, n_actions)

    # Called with either one element to determine next action, or a batch
    # during optimization. Returns tensor([[left0exp,right0exp]...]).
    def forward(self, x):
        x = F.relu(self.bn1(self.layer1(x)))
        x = self.layer1d(x)
        x = F.relu(self.bn2(self.layer2(x)))
        x = self.layer2d(x)
        x = F.relu(self.bn3(self.layer3(x)))
        x = self.layer3d(x)
        return self.layer4(x)
```

```python
class RandomDataset(torch.utils.data.Dataset):
    """
    Our random dataset
    """

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        # print(idx)
        # return {
        #     "sample": torch.tensor(self.x.iloc[idx, :], dtype=torch.float),
        #     "target": self.y.iloc[idx],
        # }
        return torch.tensor(self.x.iloc[idx, :], dtype=torch.float), self.y.iloc[idx]
```

```python
train_set = RandomDataset(X_train, y_train)
test_set = RandomDataset(X_test, y_test)

train_loader = torch.utils.data.DataLoader(train_set, batch_size=32)
test_loader = torch.utils.data.DataLoader(test_set, batch_size=32)
```

```python
model = Model(102, 4)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=1e-2)
```

```python
import matplotlib
import matplotlib.pyplot as plt

# set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display

plt.ion()
```

```python
steps = 0
trainLossX = []
trainLossY = []
testX = []
testLossY = []
accuracyY = []

def plot_loss(show_result=False):
    fig = plt.figure(1)
    # rewards_t = torch.tensor(episode_total_rewards, dtype=torch.float)
    if show_result:
        # fig.set_title('Result')
        pass
    else:
        fig.clear()
        # fig.set_title('Training...')

    ax1 = fig.subplots()
    if show_result:
        ax1.set_title('Result')
    else:
        # fig.clear()
        ax1.set_title('Training...')

    ax1.set_xlabel('Step')
    ax1.set_ylabel('Loss', color='tab:red')
    # plt.ylim((-100, 100))
    ax1.plot(trainLossX, trainLossY, color='tab:red')
    ax1.plot(testX, testLossY, color='tab:orange')
    ax1.tick_params(axis='y', labelcolor='tab:red')

    ax2 = ax1.twinx()
    ax2.set_ylabel('Accuracy', color='tab:blue')
    ax2.set_ylim((0, 1))
    ax2.plot(testX, accuracyY, color='tab:blue')
    ax2.tick_params(axis='y', labelcolor='tab:blue')

    fig.tight_layout()
    # # Take 100 episode averages and plot them too
    # if len(rewards_t) >= 100:
    #     means = rewards_t.unfold(0, 100, 1).mean(1).view(-1)
    #     means = torch.cat((torch.zeros(99), means))
    #     plt.plot(means.numpy())

    plt.pause(0.001)  # pause a bit so that plots are updated
    if is_ipython:
        if not show_result:
            # display.display(plt.gcf())
            display.display(fig)
            display.clear_output(wait=True)
        else:
            # display.display(plt.gcf())
            display.display(fig)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    global steps

    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)

        # Compute prediction error
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        steps += 1
        trainLossX.append(steps)
        trainLossY.append(loss.item())

        if batch % 100 == 0:
            plot_loss()
            # loss, current = loss.item(), (batch + 1) * len(X)
            # print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test(dataloader, model, loss_fn):
    global steps
    
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()
    test_loss /= num_batches
    correct /= size
    testX.append(steps)
    testLossY.append(test_loss)
    accuracyY.append(correct)
    plot_loss()
    # print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
epochs = 50
for t in range(epochs):
    # print(f"Epoch {t+1}\n-------------------------------")
    train(train_loader, model, loss_fn, optimizer)
    test(test_loader, model, loss_fn)
print("Done!")
plot_loss(True)
```
